```python
import pickle, numpy as np, pandas as pd
from sklearn.model_selection import StratifiedKFold,GroupKFold,KFold,train_test_split
from sklearn.preprocessing import LabelEncoder


def generate_cv_splits(graph_dataset,out_pkl,f=0.7,n_split=5,y=False):
    datasets=pickle.load(open(graph_dataset,'rb'))
    IDs=datasets['df']['ID'].unique()
    if "component" in datasets['df'].columns:
        IDs=(datasets['df']['ID']+"_"+datasets['df']['component'].astype(str)).unique()
    idx=np.arange(len(datasets['graph_dataset']))
    y=np.array([x.y.float().numpy().mean()<0.95 for x in datasets['graph_dataset']]).astype(int) if y else None
    group=LabelEncoder().fit_transform(np.array([x.id for x in datasets['graph_dataset']]))
    X=pd.DataFrame(dict(idx=idx))
    cv_splits={}#random_state=42,, shuffle=True
    np.random.seed(42)
    for i,(train_index, test_index) in enumerate((GroupKFold if not isinstance(y,type(None)) else KFold)(n_splits=n_split).split(X['idx'],y,group)):
        cv_splits[i]=dict(train_idx=train_index,
                         val_idx=test_index)

#     for i in range(len(cv_splits)):
#         new_i=i+1
#         if new_i==len(cv_splits):
#             new_i=0
#         X2=pd.DataFrame(dict(idx=cv_splits[new_i]['test_idx'],y=y[cv_splits[new_i]['test_idx']] if not isinstance(y,type(None)) else None))
#         train_index, val_index = list(GroupKFold(n_splits=n_split).split(X2['idx'], X2['y'], group[X2['idx'].values]))[0]#random_state=42, shuffle=True
#         cv_splits[i]['val_idx']=X2['idx'].values[val_index]#train_test_split(X2['idx'],train_size=f,stratify=None if isinstance(y,type(None)) else y[X2['idx'].values])[0].values
#         cv_splits[i]['train_idx']=X2['idx'].values[train_index]#np.setdiff1d(cv_splits[i]['train_idx'],cv_splits[i]['val_idx'])
    pickle.dump(cv_splits,open(out_pkl,'wb'))
```

```python
for k in ["bcc","complete"]:
    generate_cv_splits(f"{k}/graph_datasets/imagenet_graph_data.pkl",f"{k}/graph_datasets/cv_splits.pkl",f=0.7,n_split=10,y=None)
```
